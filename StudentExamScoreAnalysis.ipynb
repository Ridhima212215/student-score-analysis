```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
students = [f"Student_{i}" for i in range(1, 51)]

math_scores = np.random.randint(30, 100, size=50)
science_scores = np.random.randint(30, 100, size=50)
english_scores = np.random.randint(30, 100, size=50)
```

```python
data = {
    "Name": students,
    "Math": math_scores,
    "Science": science_scores,
    "English": english_scores
}

df = pd.DataFrame(data)
df.head()
```

```python
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
df["Pass"] = np.where((df[["Math", "Science", "English"]] >=40).all(axis=1), "Yes", "No")
```

```python
print("Top 5 students:\n", df.sort_values(by="Average", ascending=False).head(5))
print("\nHighest in Math:", df.loc[df["Math"].idxmax(), "Name"])
print("\nLowest in Science:", df.loc[df["Science"].idxmin(), "Name"])
print("Highesr in ENglish:", df.loc[df["English"].idxmax(), "Name"])
print("\nPass Count:\n", df["Pass"].value_counts())
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["Math"], alpha=0.5, label="Math")
plt.hist(df["Science"], alpha=0.5, label="Science") # alpha represents transparency 0-fully transparent 1-solid
plt.hist(df["English"], alpha=0.5, label="English")
plt.legend() #legend plot is the small box in the plt telling which color is what
plt.title("Score Distribution by Subject")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
top5 = df.sort_values(by="Average", ascending=False).head(5)
plt.figure(figsize=(10,5))
plt.bar(top5["Name"], top5["Average"], color="blue")
plt.title("Top 5 Students by Average Score")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.show()
```
